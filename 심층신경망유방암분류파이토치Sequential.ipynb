```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
import numpy as np
import torch
import torch.optim as optim
import torch.nn as nn
import random
```

```python
random_seed=178
torch.manual_seed(random_seed)
torch.backends.cudnn.deterministic=True
torch.backends.cudnn.benchmark=False
np.random.seed(random_seed)
```

```python
data=datasets.load_breast_cancer()
data
```

```python
X=data.data
X
```

```python
y=data.target
y
```

```python
y=y.reshape(-1,1)
y
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=157)
```

```python
X_train
```

```python
X_test
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(X_train)
```

```python
X_train=scaler.transform(X_train)
X_train
```

```python
X_test=scaler.transform(X_test)
X_test
```

```python
X_train=torch.FloatTensor(X_train)
X_train
```

```python
X_test=torch.FloatTensor(X_test)
X_test
```

```python
y_train=torch.FloatTensor(y_train)
y_train
```

```python
y_test=torch.FloatTensor(y_test)
y_test
```

```python
mlp_model=nn.Sequential(
    nn.Linear(in_features=30,out_features=50),
    nn.Sigmoid(),
    nn.Linear(in_features=50,out_features=10),
    nn.Sigmoid(),
    nn.Linear(in_features=10,out_features=1),
    nn.Sigmoid()
)
```

```python
mlp_model
```

```python
optimizer=optim.Adam(mlp_model.parameters(),lr=0.01)
```

```python
for i in range(1000):
    print("i=",i)
    hypothesis=mlp_model(X_train)
    print("hypothesis[:3]=",hypothesis[:3])
    cost=torch.mean(-y_train*torch.log(hypothesis)-(1-y_train)*torch.log(1-hypothesis))
    print("cost=",cost)
    optimizer.zero_grad()
    cost.backward()
    optimizer.step()
    print("="*100)
```
