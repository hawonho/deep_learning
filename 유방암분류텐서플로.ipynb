```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
import tensorflow as tf
```

```python
data=datasets.load_breast_cancer()
```

```python
data
```

```python
X=data.data
```

```python
X
```

```python
y=data.target
```

```python
y
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
X_train
```

```python
X_test
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train=scaler.transform(X_train)
```

```python
X_train
```

```python
model=Sequential()
```

```python
model.add(Dense(1,input_dim=30,activation="sigmoid"))
```

```python
model.summary()
```

```python
model.compile(loss='binary_crossentropy',optimizer=Adam(lr=0.001),metrics=['acc'])
```

```python
model.fit(X_train,y_train,epochs=1000)
```

```python
X_test=scaler.transform(X_test)
```

```python
X_test
```

```python
pred=model.predict(X_test)
```

```python
pred
```

```python
predict01=np.where(pred>0.5,1,0)
print("========================================")
print("predict01")
print(predict01)
```

```python
predict02=predict01.flatten()
```

```python
predict02
```

```python
predict03=(predict02==y_test)
```

```python
predict03
```
