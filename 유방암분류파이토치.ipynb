```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
import numpy as np
import torch
import torch.optim as optim
import torch.nn as nn
```

```python
data=datasets.load_breast_cancer()
```

```python
data
```

```python
X=data.data
```

```python
X
```

```python
y=data.target
```

```python
y
```

```python
y=y.reshape(-1,1)
```

```python
y
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
X_train
```

```python
X_test
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(X_train)
```

```python
X_train=scaler.transform(X_train)
X_train
```

```python
X_test=scaler.transform(X_test)
X_test
```

```python
w=torch.empty(30,1)
w
```

```python
w=torch.nn.init.xavier_uniform_(w)
w
```

```python
w=torch.tensor(w,requires_grad=True)
w
```

```python
b=torch.tensor([0.0],requires_grad=True,dtype=torch.float)
b
```

```python
X_train=torch.FloatTensor(X_train)
X_train
```

```python
X_test=torch.FloatTensor(X_test)
X_test
```

```python
y_train=torch.FloatTensor(y_train)
y_train
```

```python
y_test=torch.FloatTensor(y_test)
y_test
```

```python
torch.matmul(X_train,w)
```

```python
torch.matmul(X_train,w)+b
```

```python
torch.sigmoid(torch.matmul(X_train,w)+b)
```

```python
hypothesis=torch.sigmoid(torch.matmul(X_train,w)+b)
hypothesis
```

```python
optimizer=optim.Adam([w,b],lr=0.01)
```

```python
for i in range(100):
    print("i=",i)
    print("w[:3]=",w[3])
    print("b=",b)
    hypothesis=torch.sigmoid(torch.matmul(X_train,w)+b)
    cost=torch.mean(-y_train*torch.log(hypothesis)-(1-y_train)*torch.log(1-hypothesis))
    print("cost=",cost)
    optimizer.zero_grad()
    cost.backward()
    optimizer.step()
    print("="*100)
```
